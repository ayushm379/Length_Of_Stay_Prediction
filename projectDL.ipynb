```
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```
cd /content/drive/My Drive/Colab Notebooks/dataset/
```

```
data = pd.read_csv('data.csv')
```

```
Y = data['LOS'].values
X = data.drop(columns=['LOS'])
```

```
# splitting the data into training and testing datasets
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.20)
```

```
import tensorflow as tf
```

```
print(X_train.shape)
rows = X_train.shape[0]
cols = X_train.shape[1]
shape = (cols,)
```

```
from tensorflow.keras import models, layers, backend, optimizers, losses, metrics, regularizers
```

```
# creating model
model = models.Sequential()
```

```
# Input layers
model.add(layers.Dense(1024, activation='relu', input_shape = shape))
```

```
model.add(layers.Dense(1024, activation='relu', kernel_regularizer=regularizers.l2(0.001)))
model.add(layers.Dropout(0.2))
```

```
model.add(layers.Dense(1024, activation='relu', kernel_regularizer=regularizers.l2(0.001)))
model.add(layers.Dropout(0.2))
```

```
model.add(layers.Dense(1024, activation='relu', kernel_regularizer=regularizers.l2(0.001)))
model.add(layers.Dropout(0.2))
```

```
# Output layers
model.add(layers.Dense(1, activation='relu'))
print(model.summary())
```

```
# compiling model
model.compile(optimizer=optimizers.Adam(lr=1e-4), loss='mean_absolute_error', metrics=['mae'])
```

```
# fitting the model
NumEpochs = 10
batch = 8
history = model.fit(X_train, Y_train, epochs=NumEpochs, batch_size = batch, validation_data=(X_test, Y_test))
```

```
# Predicting data
results = model.evaluate(X_test, Y_test)
print('Result = {}'.format(results))
```
